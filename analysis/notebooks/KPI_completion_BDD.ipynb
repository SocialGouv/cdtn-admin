# Taux de complétion (via BDD)

```python
import pandas as pd
from analysis.connectors.matomo import MatomoSQLConnector

matomo = MatomoSQLConnector()
await matomo.connect()
```

```python
# - 10 juin / now
interval_start = '2026-06-01 00:00:00'
interval_stop = '2026-06-02 00:00:00'
```

```python
columns = ['action_id',
    'idvisit',
    'actions',
    'operatingsystemname',
    'action_type',
    'action_eventcategory',
    'action_eventaction',
    'action_eventname',
    'action_eventvalue',
    'action_url',
    'experiments']

range_query = f"""
        SELECT {", ".join(columns)} FROM matomo_partitioned
        WHERE action_timestamp >= '{interval_start}'
          AND action_timestamp < '{interval_stop}'
        """
```

```python
def normalize(name) -> str:
    # décode + minuscule + trim pour comparer proprement
    return unquote(str(name)).strip().lower()

mapping_df["filename_norm"] = mapping_df["filename"].map(normalize)

# 2. Isoler les downloads
downloads_df = visits_df[visits_df["action_type"] == "download"].copy()

# 3. Extraire le nom de fichier depuis action_url (dernier segment de path, décodé)
def extract_filename(url) -> str:
    if not isinstance(url, str):
        return ""
    path = urlparse(url).path            # enlève domaine + query string
    last = path.rsplit("/", 1)[-1]       # dernier segment
    return normalize(last)

downloads_df["filename_norm"] = downloads_df["action_url"].map(extract_filename)

# 4. Rattacher chaque download à un slug
downloads_df = downloads_df.merge(
    mapping_df[["slug", "filename_norm"]],
    on="filename_norm",
    how="left",
)

# 5. Nombre de visites uniques ayant téléchargé, par modèle
downloads_per_model = (
    downloads_df.dropna(subset=["slug"])
    .groupby("slug")["idvisit"]
    .nunique()
    .rename("nb_downloads")
    .reset_index()
    .sort_values("nb_downloads", ascending=False)
)
```

```python
# Slug normalisé pour matcher les segments d'URL
mapping_df["slug_lower"] = mapping_df["slug"].str.lower()

# 1. Isoler les vues de page
pages_df = visits_df[visits_df["action_type"] == "action"].copy()

# 2. Découper action_url en segments de path (décodés, minuscule)
def url_segments(url):
    if not isinstance(url, str):
        return []
    path = unquote(urlparse(url).path).strip("/").lower()
    return path.split("/") if path else []

pages_df["segments"] = pages_df["action_url"].map(url_segments)

# 3. Exploser les segments et ne garder que ceux qui sont des slugs de modèle
model_pageviews = (
    pages_df[["idvisit", "action_url", "segments"]]
    .explode("segments")
    .rename(columns={"segments": "slug_lower"})
    .merge(mapping_df[["slug", "slug_lower"]], on="slug_lower", how="inner")
)

# 4. Nombre de vues par modèle
pageviews_per_model = (
    model_pageviews.groupby("slug")["idvisit"]
    .nunique()
    .rename("nb_pageviews")
    .reset_index()
)
```

```python
taux = (
    pageviews_per_model
    .merge(downloads_per_model, on="slug", how="left")
    .assign(nb_downloads=lambda d: d["nb_downloads"].fillna(0).astype(int))
)
taux["taux_download"] = taux["nb_downloads"] / taux["nb_pageviews"]
taux = taux.sort_values("taux_download", ascending=False)
taux
```
